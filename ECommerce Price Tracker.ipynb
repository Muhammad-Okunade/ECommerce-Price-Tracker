```python
def price_tracker() :
    url = 'https://www.jumia.com.ng/samsung-samsung-galaxy-a12-6.5-48mp-camera-464gb-memory-5000mah-battery-fingerprint-4g-lte-red-81673047.html'
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36 Edg/96.0.1054.62"} 
    
    page = requests.get(url, headers = headers) 
    soup1 = BeautifulSoup (page.content, 'html.parser')
    soup2 = BeautifulSoup (soup1.prettify(), 'html.parser')
    
    title = soup2.find(class_ = '-fs20 -pts -pbxs').get_text()
    price = soup2.find(class_="-hr -mtxs -pvs").get_text()
    brand = soup2.find(href="/samsung/").get_text()
    rating = soup2.find(class_="stars _s _al").get_text()
    timestamp = datetime.datetime.today()
    
    title = title.strip()
    price = price.strip()[2:]
    brand = brand.strip()
    rating = rating.strip()[:3]
    
    header = ['Title', 'Price(NGN)', 'Brand', 'Rating', 'Last Updated']
    data = [title, price, brand, rating, timestamp]

    with open ('JumiaPhonesList.csv', 'a+', newline = '', encoding = 'UTF8') as f:
        writer = csv.writer(f)
        writer.writerow(data)
```

```python
while(True):
    price_tracker()
    time.sleep(600) #Updates every 10mins
```
